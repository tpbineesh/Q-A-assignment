```python
##Check whether all the zones generate the same amount of sales.
Null hypothesis- there is no difference in sales generated in each zones
Alternate hypothesis- there is difference in sales generated in each zones
```

```python
import scipy.stats as stats
# stats f_oneway functions takes the zones as input and returns ANOVA F and p value
fvalue, pvalue = stats.f_oneway(sales_data['Zone - A' ],sales_data['Zone - B'],sales_data['Zone - C'],sales_data['Zone - D'])
print(fvalue, pvalue)
```

since p value is less than 0.05, there is significant difference between sales in all three zones.
