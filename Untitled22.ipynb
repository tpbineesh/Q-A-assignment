```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, auc, roc_curve
```

The dataset look balanced

```python
dataset = dataset.astype('category')
dataset.dtypes
```

```python
#These are categorical values. Using label encoder to convert it to ordinal
```

```python
labelencoder=LabelEncoder()
for column in dataset.columns:
    dataset[column] = labelencoder.fit_transform(dataset[column])
```

```python
dataset.describe()
```

```python
Here veil-type is not contributing to dataset. Thus dropping it
```

```python
dataset2=dataset.drop(['veil-type'], axis=1)
dataset2
```

```python
# Plotting heatmap to find correlation
```

```python
plt.figure(figsize=(14,12))
sns.heatmap(dataset2.corr(),linewidths=.1,cmap="Purples", annot=True, annot_kws={"size": 7})
plt.yticks(rotation=0);
```

```python
#Preparing data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = dataset2.drop(['class'], axis=1)
```

```python
y=dataset2['class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
#Use logistic regression classification
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(solver="lbfgs", max_iter=500)
lr.fit(x_train, y_train)
print("Test Accuracy: {}%".format(round(lr.score(x_test, y_test)*100,2)))
```

```python
#classification report
```

```python
y_pred_lr = lr.predict(x_test)
print("Logistic Regression Classifier report: \n\n", classification_report(y_test, y_pred_lr))
```

```python
#Confusion matrice
```

```python
import matplotlib.pyplot as plt 
```

```python
cm = confusion_matrix(y_test, y_pred_lr)
x_axis_labels = ["Edible", "Poisonous"]
y_axis_labels = ["Edible", "Poisonous"]
```

```python
f, ax = plt.subplots(figsize =(7,7))
sns.heatmap(cm, annot = True, linewidths=0.2, linecolor="black", fmt = ".0f", ax=ax, cmap="Purples", xticklabels=x_axis_labels, yticklabels=y_axis_labels)
```

```python
#Naive Bayes Classification
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
print("Test Accuracy: {}%".format(round(nb.score(x_test, y_test)*100, 2)))
```

Classification report

```python
y_pred_nb = nb.predict(x_test)
```

```python
print("Naive Bayes Classifier report: \n\n", classification_report(y_test, y_pred_nb))
```

```python
y_pred=nb.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)
cm
```
